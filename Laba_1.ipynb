```python
import pandas as pd
import numpy as np
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
data_users = pd.read_table("D:\\Projects\\IDA\\ml-1m\\users.dat", sep= "::", engine='python', names=unames)
data_users
```

```python
unames = ['movie_id', 'title', 'genre']
data_movies = pd.read_table("D:\\Projects\\IDA\\ml-1m\\movies.dat", sep= "::", engine='python', encoding='ISO-8859-1', names=unames)
data_movies
```

```python
unames = ['user_id', 'movie_id', 'rating', 'timestamp']
data_ratings = pd.read_table("D:\\Projects\\IDA\\ml-1m\\ratings.dat", sep= "::", engine='python', names=unames)
data_ratings
```

```python
df1 = pd.DataFrame(data_users)
df2 = pd.DataFrame(data_movies)
df3 = pd.DataFrame(data_ratings)
df4 = df3.merge(df2, on='movie_id')
df5 = df4.merge(df1, on='user_id')
df5
```

```python
df6 = df5.drop(columns=['timestamp', 'zip'])
df6
```

```python
df7 = df6
bins = [0, 20, 50, np.inf]
category = ['young', 'adult', 'old']
df7['experience'] = pd.cut(df7['age'], bins, labels=category)
df7
```

```python
df7[(df7['experience'] == 'young') & (df7['gender'] == 'M')].sort_values('rating', ascending=False).head(10)
```

```python
df8 = df7[df7['gender'] == 'M']
df8[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df9 = df7[df7['gender'] == 'F']
df9[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df10 = df7[(df7['experience'] == 'young') & (df7['gender'] == 'M')]
df10[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df11 = df7[(df7['experience'] == 'young') & (df7['gender'] == 'F')]
df11[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df12 = df7[(df7['experience'] == 'adult') & (df7['gender'] == 'F')]
df12[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df13 = df7[(df7['experience'] == 'adult') & (df7['gender'] == 'M')]
df13[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df14 = df7[(df7['experience'] == 'old') & (df7['gender'] == 'M')]
df14[['title', 'rating', 'gender']].groupby(['title', 'gender']).sum().sort_values(by = 'rating', ascending=False).head(10)
```

```python
df15 = df7[(df7['experience'] == 'old') & (df7['gender'] == 'F')]
top_10 = df15[['title', 'rating']].groupby(['title']).sum().sort_values(by = 'rating', ascending=False).head(10)
ax = top_10.plot(kind='barh', title='Top-10 movies for old people', ylabel='Rating', xlabel='', figsize=(6, 5))
```
