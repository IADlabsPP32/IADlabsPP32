```python
import pandas as pd
import matplotlib.pyplot as plt

unames = ['UserId', 'Gender', 'Age', 'Occupation', 'Zip-code']
data_users = pd.read_table(".\\ml-1m\\users.dat", sep= "::", engine='python', names=unames)

unames=['UserID','MovieID','Rating','Timestamp']
data_ratings=pd.read_table(".\\ml-1m\\ratings.dat", sep= "::", engine='python', names=unames)

unames=['MovieID','Title','Genres']
data_movies=pd.read_table(".\\ml-1m\\movies.dat", sep="::", engine='python', names=unames, encoding='ISO-8859-1')

data_users_ratings=data_ratings.merge(data_users, left_on="UserID", right_on="UserId")
data_merged=data_users_ratings.merge(data_movies, left_on="MovieID", right_on="MovieID")

data_age_young=data_merged.query("Age>=1 & Age<18")
data_age_adult=data_merged.query("Age>=18 & Age<35")
data_age_midle=data_merged.query("Age>=35 & Age<50")
data_age_old=data_merged.query("Age>=50")
data_age_category_56=data_merged.query("Age>=56")


data_male=data_merged[data_merged['Gender']=='M'].sort_values('Rating', ascending=False)
data_female=data_merged[data_merged['Gender']=='F'].sort_values('Rating', ascending=False)


data_male[['Rating','Title']].groupby('Title').sum().sort_values(by='Rating', ascending=False).head(10)
data_female[['Rating','Title']].groupby('Title').sum().sort_values(by='Rating', ascending=False).head(10)

```
