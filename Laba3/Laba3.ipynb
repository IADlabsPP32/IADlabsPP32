```python
cov = np.zeros(18)
a = 0
for i in np.arange(4, 22):
    fac = Titles[i]
    cov[a] = data['Life expectancy '].cov(data[fac])
    print("коваріація для ",Titles[i],":",cov[a])
    a = a + 1
```

```python
corv = np.zeros(18)
a = 0
for i in np.arange(4, 22):
    fac = Titles[i]
    corv[a] = data['Life expectancy '].corr(data[fac])
    print("кореляція для ",Titles[i],":",corv[a])
    a = a + 1
```

```python
disp = np.zeros(18)
a = 0
for i in np.arange(4, 22):
    fac = Titles[i]
    disp[a] = data[fac].var()
    print("дисперсія для ",Titles[i],":",disp[a])
    a = a + 1
```

```python
allData = zip(Titles, cov, corv, disp)
rezCol = ['Фактор','Коваріація','Кореляція','Дисперсія']
Rez = pd.DataFrame(data=allData, columns = rezCol)
Rez
```

```python
factors = ['Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles ',' BMI ','under-five deaths ','Polio','Total expenditure','Diphtheria ',' HIV/AIDS','GDP','Population',' thinness  1-19 years',' thinness 5-9 years','Income composition of resources','Schooling']

firstUsedFactors = ['Adult Mortality','infant deaths','Alcohol',' BMI ','under-five deaths ','Population']
print("Дані, що були обрані спочатку")
for i in np.arange(0,6):
    result = data[firstUsedFactors[i]].corr(data['Life expectancy '])
    if(-0.09<=result<=0 or 0<= result<=0.09): print(firstUsedFactors[i], " Кореляція відсутня ", result)
    if(-0.3<result<=-0.1or 0.1<=result<0.3): print(firstUsedFactors[i], " Низька ", result)
    if(-0.5<result<=-0.3or 0.3<=result<0.5): print(firstUsedFactors[i], " Середня ", result)
    if(-1<=result<=-0.5 or 0.5<=result<=1): print(firstUsedFactors[i], " Висока ", result)
print(" ")
print("Дані, що мають вплив на тривалість життя")
for i in np.arange(0, 18):
    result = data[factors[i]].corr(data['Life expectancy '])
    if(-0.5<result<=-0.3or 0.3<=result<0.5): print(Titles[i], " Середня ", result)
    if(-1<=result<=-0.5 or 0.5<=result<=1): print(Titles[i], " Висока ", result)
```

```python
print('Чи повинна країна з нижчим значенням очікуваної тривалості життя (<65) збільшити витрати на охорону здоров’я, щоб покращити середню тривалість життя?')

under65 = data[data['Life expectancy ']<65]
under65s = under65.sample(n = 20)
fig, ax = plt.subplots()
plt.title("Залежність очікуваної тривалості життя від витрат на охорону здоров'я")
ax.set_xticks(np.arange(30, 65))
ax = sb.scatterplot(under65s, x='Life expectancy ', y ='Total expenditure', hue = 'Country', )
sb.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()

under65cor = under65['Total expenditure'].corr(under65['Life expectancy '])
print("Витрати на ОЗ мають залежність від очікуваної тривалості життя:")
if(-0.09<=under65cor<=0 or 0<= under65cor<=0.09): print(" Кореляція відсутня ", under65cor)
if(-0.3<under65cor<=-0.1or 0.1<=under65cor<0.3): print(" Низьку ", under65cor)
if(-0.5<under65cor<=-0.3or 0.3<=under65cor<0.5): print( " Середню ", under65cor)
if(-1<=under65cor<=-0.5 or 0.5<=under65cor<=1): print( " Високу ", under65cor)
```

```python
print('Чи має очікувана тривалість життя позитивну чи негативну кореляцію зі звичками в їжі, способом життя, фізичними вправами, курінням, вживанням алкоголю тощо.')
print(' ')
habits = data
result = habits['Life expectancy '].corr(habits['Alcohol'])
if(result>=0): print("Alcohol має позитивну кореляцію")
else: print("Alcohol має негативну кореляцію")

result = habits['Life expectancy '].corr(habits[' BMI '])
if(result>=0): print("BMI має позитивну кореляцію")
else: print("BMI має негативну кореляцію")
```

```python
print('Який вплив навчання в школі на тривалість життя людини?')

schooling = data.sample(n = 30)
fig, ax= plt.subplots()
ax.set_xticks(np.arange(30, 100))
plt.xticks(rotation=45)
plt.title("Залежність тривалість життя людини від навчання у школі")
ax = sb.scatterplot(schooling, x='Life expectancy ', y ='Schooling', hue = 'Country')
sb.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()

print("Навчанняу школі має ")
result = data['Life expectancy '].corr(data['Schooling'])
if(-0.09<=result<=0 or 0<= result<=0.09): print(" не має залежності ", result)
if(-0.3<result<=-0.1or 0.1<=result<0.3): print(" низьку залежність ", result)
if(-0.5<result<=-0.3or 0.3<=result<0.5): print(" середню залежність ", result)
if(-1<=result<=-0.5 or 0.5<=result<=1): print(" високу залежність ", result)
```

```python
print('Позитивний чи негативний зв’язок очікуваної тривалості життя з вживанням алкоголю?')
habits = data
result = habits['Life expectancy '].corr(habits['Alcohol'])
if(result>=0): print("Alcohol має позитивний зв'язок до очікуваної тривалості життя")
else: print("Alcohol має негативний зв'язок до очікуваної тривалості життя")
```
