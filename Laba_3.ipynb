```python
df1 = df.drop(columns=['Year', 'Income composition of resources', 'Schooling', 'percentage expenditure', 'under-five deaths '])
df2 = df1.corr()
plt.figure(figsize=(16,12))

# Create a custom diverging palette
cmap = sns.diverging_palette(250, 15, s=75, l=40, n=9, center="light", as_cmap=True)
_ = sns.heatmap(df2, center=0, annot=True, fmt='.2f', square=True, cmap=cmap)
```

```python
cols = ['Hepatitis B', 'Polio', 'Diphtheria', 'BMI'] 
dfc = df[["Hepatitis B", "Polio", "Diphtheria ", " BMI "]]
dfc = dfc.cov()
plt.figure(figsize=(8,8))
plt.title('This matrix showing covariance and variance coefficients', size = 16)
sns.set(font_scale=1.5)
hm = sns.heatmap(dfc,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 12},
                 cmap='coolwarm',                 
                 yticklabels=cols,
                 xticklabels=cols)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="darkgrid")
sns.histplot(data=df1, x="Hepatitis B", color="skyblue", label="Hepatitis B", kde=True)
sns.histplot(data=df1, x="Polio", color="red", label="Polio", kde=True)
sns.histplot(data=df1, x="Diphtheria ", color="gold", label="Diphtheria", kde=True)
plt.legend() 
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(7, 7)) 
sns.histplot(data=df1, x="Hepatitis B", kde=True, color="skyblue", bins = 25, ax=axs[0, 0])
sns.histplot(data=df1, x="Polio", kde=True, color="olive", bins = 25, ax=axs[0, 1])
sns.histplot(data=df1, x="Diphtheria ", kde=True, color="gold", bins = 25, ax=axs[1, 0])
sns.histplot(data=df1, x=" thinness  1-19 years", kde=True, color="teal", bins = 25, ax=axs[1, 1])
plt.show()
```
