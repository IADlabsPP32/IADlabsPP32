```python
import pandas as pd

movies = ['MovieID', 'Title', 'Genres']
ratings = ['UserID', 'MovieID', 'Rating', 'Timestamp']
users = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']

data_movies = pd.read_table(".\\ml-1m\\movies.dat", sep="::", engine='python', names=movies)
moviesData = pd.DataFrame(data_movies, columns=movies)

data_ratings = pd.read_table(".\\ml-1m\\ratings.dat", sep="::", engine='python', names=ratings)
ratingsData = pd.DataFrame(data_ratings, columns=ratings)

data_users = pd.read_table(".\\ml-1m\\users.dat", sep="::", engine='python', names=users)
usersData = pd.DataFrame(data_users, columns=users)

mergedData = moviesData.merge(ratingsData.merge(usersData))

under_18_Group = mergedData[mergedData['Age'] == 1]
group_18_24 = mergedData[mergedData['Age'] == 18]
group_25_34 = mergedData[mergedData['Age'] == 25]
group_35_44 = mergedData[mergedData['Age'] == 35]
group_45_49 = mergedData[mergedData['Age'] == 45]
group_50_55 = mergedData[mergedData['Age'] == 50]
older_60_Group = mergedData[mergedData['Age'] == 60]

groupByMale = mergedData[mergedData['Gender'] == 'M']
groupByFemale = mergedData[mergedData['Gender'] == 'F']

print(under_18_Group.sort_values(by=['Rating'], ascending=False).head(10))
print(group_18_24.sort_values(by=['Rating'], ascending=False).head(10))
print(group_25_34.sort_values(by=['Rating'], ascending=False).head(10))
print(group_35_44.sort_values(by=['Rating'], ascending=False).head(10))
print(group_45_49.sort_values(by=['Rating'], ascending=False).head(10))
print(group_50_55.sort_values(by=['Rating'], ascending=False).head(10))
print(older_60_Group.sort_values(by=['Rating'], ascending=False).head(10))
print(groupByMale.sort_values(by=['Rating'], ascending=False).head(10))
print(groupByMale.sort_values(by=['Rating'], ascending=False).head(10))
```
