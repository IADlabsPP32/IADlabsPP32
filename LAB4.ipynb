```python
countries_new = countries_data[countries_data['gdpp'] < 37000]
countries_new = countries_data[countries_data['child_mort'] < 140]
countries_new = countries_data[countries_data['income'] < 50000]
countries_new.nlargest(10, 'income')
```

```python
cluster_ = list(zip(countries_new["gdpp"], countries_new["income"]))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(cluster_)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), inertias, marker='x', color = 'orange')
plt.title('Метод ліктя')
plt.xlabel('Кількість кластерів')
plt.ylabel('Інерція')
plt.show()
```

```python
countries_new.nsmallest(10, ['income', 'gdpp'])
```

```python
CI = countries_new[['child_mort', 'income']].copy()
from sklearn.mixture import GaussianMixture
n_clusters = 3
gmm_model = GaussianMixture(n_components=n_clusters)
gmm_model.fit(CI)
cluster_labels = gmm_model.predict(CI)
CI = pd.DataFrame(CI)
CI['cluster'] = cluster_labels
fig, ax = plt.subplots(figsize=(9, 7))
sns.scatterplot(x='child_mort', y='income', data=CI, hue='cluster', cmap='Pastel2', s=70, ax = ax) 
plt.xlabel('income')
```

```python
worst_values = countries_new[["country"]]
worst_values=pd.merge(worst_values,CI, left_index=True,right_index=True)
worst_values.nsmallest(5, ['income', 'child_mort'])
```

```python
worst_values['cluster'].value_counts(normalize=True)
```

```python
cluster_childmort=pd.DataFrame(CI.groupby(["cluster"]).child_mort.mean())
cluster_income=pd.DataFrame(CI.groupby(["cluster"]).income.mean())
CMI = pd.concat([cluster_childmort,cluster_income], axis=1)
CMI.columns = ["child_mort","income"]
CMI
```

```python
fig = plt.figure(figsize = (10,6))
CMI.rename(index={2: 'Високо розвинуті країни'},inplace=True)
CMI.rename(index={1: 'Країни, що розвиваються'},inplace=True)
CMI.rename(index={0: 'Низько розвинуті країни'},inplace=True)
s=sns.barplot(x=CMI.index,y='income',data=CMI)
plt.xlabel('Groups', fontsize=10)
plt.ylabel('Income per person', fontsize=10)
plt.title('Group countries by income')
plt.show()
```
