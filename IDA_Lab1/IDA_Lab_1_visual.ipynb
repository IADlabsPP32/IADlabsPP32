```python
import pandas as pd;

import seaborn as sns;
import matplotlib.pyplot as plt
movies = ['MovieID', 'Title', 'Genres'];
data_movies = pd.read_table(".\\ml-1m\\movies.dat", sep="::", engine='python', names=movies);
moviesDataFrame = pd.DataFrame(data_movies, columns=movies);

ratings = ['UserID', 'MovieID', 'Rating', 'Timestamp'];
data_ratings = pd.read_table(".\\ml-1m\\ratings.dat", sep="::", engine='python', names=ratings);
ratingsDataFrame = pd.DataFrame(data_ratings, columns=ratings);

users = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
data_users = pd.read_table(".\\ml-1m\\users.dat", sep="::", engine='python', names=users);
usersDataFrame = pd.DataFrame(data_users, columns=users);

mergedDataFrame = ratingsDataFrame.merge(usersDataFrame).merge(moviesDataFrame);

under_18_Group = mergedDataFrame[mergedDataFrame['Age'] == 1];
between_18_24_Group = mergedDataFrame[mergedDataFrame['Age'] == 18];
between_25_34_Group = mergedDataFrame[mergedDataFrame['Age'] == 25];
between_35_44_Group = mergedDataFrame[mergedDataFrame['Age'] == 35];
between_45_49_Group = mergedDataFrame[mergedDataFrame['Age'] == 45];
between_50_55_Group = mergedDataFrame[mergedDataFrame['Age'] == 50];
older_56_Group = mergedDataFrame[mergedDataFrame['Age'] == 56];

under_18_Male = under_18_Group[under_18_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_18_24_Male = between_18_24_Group[between_18_24_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_25_34_Male = between_25_34_Group[between_25_34_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10);
between_35_44_Male = between_35_44_Group[between_35_44_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_45_49_Male = between_45_49_Group[between_45_49_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_50_55_Male = between_50_55_Group[between_50_55_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
older_56_Male = older_56_Group[older_56_Group['Gender']=='M'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10) 

under_18_Female = under_18_Group[under_18_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_18_24_Female = between_18_24_Group[between_18_24_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_25_34_Female = between_25_34_Group[between_25_34_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10);
between_35_44_Female = between_35_44_Group[between_35_44_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_45_49_Female = between_45_49_Group[between_45_49_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
between_50_55_Female = between_50_55_Group[between_50_55_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10)
older_56_Female = older_56_Group[older_56_Group['Gender']=='F'].groupby(['Title'])[['Rating']].sum().sort_values(by=['Rating'], ascending=False).head(10) 

sns.set(style='darkgrid')

sns.scatterplot(data=under_18_Male,x='Rating',y='Title')
```
