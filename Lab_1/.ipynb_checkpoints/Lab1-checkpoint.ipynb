```python
import math 
import pandas
import seaborn as sb
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

def create_tables():
    pandas.set_option('display.max_columns', None)

    movies_content = pandas.read_table('DataSet/movies.dat', sep = '::', encoding = 'ISO-8859-1', engine = 'python')
    ratings_content = pandas.read_table('DataSet/ratings.dat', sep = '::', encoding = 'ISO-8859-1', engine = 'python')
    users_content = pandas.read_table('DataSet/users.dat', sep = '::', encoding = 'ISO-8859-1', engine = 'python')

    merged_table = ratings_content.merge(movies_content, right_on = 'MovieID', left_on = 'MovieID')
    merged_table = merged_table.merge(users_content, right_on = 'UserID', left_on = 'UserID')


    kid_raitings = merged_table.query("Age >= 1 & Age <= 11")              #range 1-11
    teen_raitings = merged_table.query("Age >= 12 & Age <= 17")            #range 12-17    
    youth_raitings = merged_table.query("Age >= 18 & Age <= 29")           #range 18-29
    adults_raitings = merged_table.query("Age >= 30 & Age <= 54")          #range 30-54
    old_raitings = merged_table.query("Age >= 55")                         #range 55-...

    male_raitings = merged_table.query("Sex == 'M'")
    female_raitings = merged_table.query("Sex == 'F'")
    
    
    
    print(f"All male:\n{male_raitings[['Rating', 'Title']].groupby('Title').sum().sort_values(by = ['Rating'], ascending = False).head(10)}\n")
    print(f"All female:\n{female_raitings[['Rating', 'Title']].groupby('Title').sum().sort_values(by = ['Rating'], ascending = False).head(10)}\n")
    
    all_ages_tables = {
        11 : kid_raitings,
        17 : teen_raitings,
        29 : youth_raitings,
        54 : adults_raitings,
        55 : old_raitings
    }
    
    index = 0
    tables_dict = {}
    for table in all_ages_tables.values() :

        age = list(all_ages_tables.keys())[index]
        index += 1

        msg = 'under' if age < 55 else 'older then'

        male_age_raitings = table.query("Sex == 'M'")
        female_age_raitings = table.query("Sex == 'F'")

        male_age_raitings = male_age_raitings[['Rating', 'Title']].groupby('Title').sum().sort_values(by = ['Rating'], ascending = False).head(10)
        female_age_raitings = female_age_raitings[['Rating', 'Title']].groupby('Title').sum().sort_values(by = ['Rating'], ascending = False).head(10)
        
        str_m = f"All male {msg} {age} years";
        print(f"{str_m}:\n{male_age_raitings}\n")
        
        str_f = f"All female {msg} {age} years";
        print(f"{str_f}:\n{female_age_raitings}\n")

        tables_dict[str_m] = male_age_raitings
        tables_dict[str_f] = female_age_raitings
        
    return tables_dict

def create_visualization(tables) :
    rows_count = math.ceil(len(tables.keys()) / 2)
    
    fig, axs = plt.subplots(rows_count, 2, figsize=(50,60))
    
    fig.tight_layout(pad=0.4, w_pad=3, h_pad=18)
    
    tables_names = list(tables.keys())

    for i in range(len(axs)):
        for j in range(len(axs[i])):
            name_idx = i * len(axs[i]) + j
            if name_idx < len(tables_names):
                name = tables_names[name_idx]
                axs[i, j].set_title(name)
                
                titles = tables[name].index.tolist()
                
                axs[i, j].set_xticks(np.arange(len(titles)))
                axs[i, j].set_xticklabels(titles)
                
                tables[name].plot(ax=axs[i, j])          
    
                plt.setp(axs[i, j].get_xticklabels(), rotation=30, fontsize=17)
                

   

if __name__ == '__main__' :
    tables = create_tables()
    create_visualization(tables)
```
