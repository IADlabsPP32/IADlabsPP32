```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df
```

```python
df.info()
```

```python
df.sample(10)
```

# Countries with largest life expectancy 

```python
df.nlargest(10, ['Life expectancy ', 'Year'])
```

# Countries with smallest life expectancy 

```python
df.nsmallest(10, ['Life expectancy ', 'Year'])
```

```python
df.shape
```

```python
country_list = df['Country'].unique()
metrics_list = df.keys()
metrics_list
```

```python

for country in country_list:
    df.loc[df['Country'] == country, metrics_list] = df.loc[df['Country'] == country,metrics_list].interpolate()
df.dropna(inplace=True)
df.isnull().sum()

```

```python
sns.set(font_scale=1.3)
data_numeric = df.select_dtypes(include='number')
column_numeric = data_numeric.columns.drop('Year')
```

# Boxplots

```python

plt.figure(figsize=(8,65))
plt.subplots_adjust(wspace=1, hspace=1)
plt.style.use('dark_background')

i = 1
for col_name in column_numeric:
    plt.subplot(20,2,i)
    sns.boxplot(data=df, y=col_name)
    i +=1
```

# Hisplots

```python

plt.figure(figsize=(10, 60))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

i = 1
for col_name in column_numeric:
    plt.subplot(20,2,i)
    sns.histplot(data = df[col_name],  bins = 25)
    sns.rugplot(data = df[col_name], height=-0.02, clip_on=False)
    i +=1
```

# Variance and Covariance

```python
df.var() #Видає попередження, але працює корректно
```

```python
covariance = df.cov()
covariance
```

# Correlation

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, cmap='PiYG')
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(method="spearman"), annot=True, cmap='PiYG')
plt.show()
```
