```python
clearDat = countriesDat[["country",'gdpp','child_mort','income']]
clearDat = clearDat[clearDat['gdpp'] < 55000]
clearDat = clearDat[clearDat['child_mort'] < 140]
clearDat = clearDat[clearDat['income'] < 50000]
clearDat.info()
```

## Clustering with k-mean method

```python
clust = list(zip(clearDat["gdpp"], clearDat["income"]))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(clust)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), inertias, marker='x', color = 'purple')
plt.title('Elbow method')
plt.xlabel('Count of clusters')
plt.ylabel('Inertion')
plt.show()
```

```python
clust2 = list(zip(clearDat["gdpp"], clearDat["child_mort"]))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(clust2)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), inertias, marker='x', color = 'purple')
plt.title('Elbow method')
plt.xlabel('Count of clusters')
plt.ylabel('Inertion')
plt.show()
```

```python
cmeans = KMeans(n_clusters=2)
kmeans.fit(clust)
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(clearDat["gdpp"], clearDat["income"], c=kmeans.labels_, cmap='PiYG')
ax.set_xlabel('gdpp')
ax.set_ylabel('income')
plt.show()
```

```python
clearDat.nsmallest(10, ['income', 'gdpp'])
```

```python
CI = clearDat[['child_mort', 'income']].copy()
from sklearn.mixture import GaussianMixture
n_clusters = 3
gmm_model = GaussianMixture(n_components=n_clusters)
gmm_model.fit(CI)
cluster_labels = gmm_model.predict(CI)
CI = pd.DataFrame(CI)
CI['cluster'] = cluster_labels
```

```python
fig, ax = plt.subplots(figsize=(9, 7))
sns.scatterplot(x='child_mort', y='income', data=CI, hue='cluster', cmap='PiYG', s=70, ax = ax) 
plt.title('Clusters Identified by Guassian Mixture Model')
plt.ylabel('child_mort')
plt.xlabel('income')
```

## 5 countries with worst values

```python
badCount = clearDat[["country"]]
badCount=pd.merge(badCount,CI, left_index=True,right_index=True)
badCount.nsmallest(5, ['income', 'child_mort'])
```

```python
badCount['cluster'].value_counts(normalize=True)
```

```python

Cluster_child_mort=pd.DataFrame(CI.groupby(["cluster"]).child_mort.mean())
Cluster_income=pd.DataFrame(CI.groupby(["cluster"]).income.mean())
CMI = pd.concat([Cluster_child_mort,Cluster_income], axis=1)
CMI.columns = ["child_mort","income"]
CMI
```

```python

fig = plt.figure(figsize = (10,6))
CMI.rename(index={2: 'High developes countries'},inplace=True)
CMI.rename(index={1: 'Developing countries'},inplace=True)
CMI.rename(index={0: 'Low developed countries'},inplace=True)
s=sns.barplot(x=CMI.index,y='income',data=CMI)
plt.xlabel('Groups', fontsize=10)
plt.ylabel('Income per person', fontsize=10)
plt.title('Group countries by income')
plt.show()
```
